```python
#sin curve
t=np.arange(0.0,2.0,0.01)
s= 1+np.sin(2*np.pi*t);
plt.plot(t,s)
plt.xlabel("Sin Curve ---->")
plt.ylabel("Time Series ---->")
plt.title("Trigometric Graph ---->")
plt.grid(True)
plt.savefig("/content/sample_data/image/physics.png")
plt.show()
```

# Bar Chart Visualization

```python
#Bar graph
y=[3,10,2,5,8]
x=range(len(y))
width = 1/2
colours = ['darkblue','blue','dodgerblue','deepskyblue','skyblue']
fig, axl = plt.subplots(2,2)
for i, ax in enumerate(axl.flat):
  if i < len(y):
    ax.bar(x,y,width = width, color=colours[i])
    ax.set_xlabel("Current --------->")
    ax.set_ylabel("Velocity --------->")
    ax.set_title(F"Bar Chart")
    ax.grid(False)
plt.tight_layout()
plt.show()
```

# Bubble Chart Visualization

```python
N=50
x=np.random.rand(N);
y=np.random.rand(N);
colors = np.random.rand(N);
area=np.pi*(15*np.random.rand(N))**2;
plt.scatter(x,y,s=area, c=colors, alpha=0.8)
plt.show()
```

# Pie Chart Visualization

```python
l = 'Data Science','Machine Learning','Deep Learning','Analytics','AI'
sizes = [15,30,45,5,10]
exp = (0.3,0,0,0,0)
fig1,ax1 = plt.subplots();
ax1.pie(sizes, explode=exp, labels = l, autopct='%1.6f%%', shadow = True, startangle=100);
ax1.axis('equal')
plt.show()
```
